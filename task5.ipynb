<h3>Lets see the graphical representation of dataset.

```python
plt.subplot(221)
data['gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(16,10))
plt.xticks(rotation=0)

plt.subplot(222)
data['race/ethnicity'].value_counts().plot(kind='bar', title='Race/ethnicity of students')
plt.xticks(rotation=0)

plt.subplot(223)
data['lunch'].value_counts().plot(kind='bar', title='Lunch status of students')
plt.xticks(rotation=0)

plt.subplot(224)
data['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')
plt.xticks(rotation=0)

plt.show()
```

<p>1. From <b>gender graph</b> we can see that school contains more number of female students.<br>
    2. From <b>ethnicity graph</b> majority of the students belong to groups C and D.<br>
    3. From <b>lunch graph</b> more than 60% of the students have a standard lunch at school.<br>
    4. From <b>test preparation</b> graphmore than 60% of students have not taken any test preparation course.

<h4>We use boxplot to identify the outliers.

```python
data.boxplot()
```

<p>Hollow circles near the tails represent outliers in the dataset.<br>
    It is very much possible for a student to score extremely low marks in a test, we will not remove these outliers.

<h4>We make distribution plot of the students math score.

```python
sns.displot(data['math score'])
```

<p>The peak is at around 65 marks, the mean of the math score of the students in the dataset.

<h4>Representation of correlation between the 3 scores with the help of a heatmap.

```python
cor = data.corr()
sns.heatmap(cor, annot = True, square = True)
plt.yticks(rotation=0)
plt.show()
```

<p>The heatmap shows that the 3 scores are highly correlated. Reading score has a correlation coefficient of 0.95 with the writing score. Math score has a correlation coefficient of 0.82 with the reading score, and 0.80 with the writing score.

<h4>Relational plot to understand the relationship between 2 variables.

```python
sns.relplot(x='math score', y='writing score', hue='gender', data=data)
```

<p>graph shows a difference in scores between the male and female students.<br>
    For the same math score, female students are more likely to have a higher writing score than male students. <br>
    For the same writing score, male students are expected to have a higher math score than female students.

<h4>Impact of the test preparation course on students performance using a horizontal bar graph.

```python
data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().T.plot(kind='barh', figsize=(10,10))
```

<p>students who have completed the test preparation course have performed better, on average, as compared to students who have not opted for the course.
